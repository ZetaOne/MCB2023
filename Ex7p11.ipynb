```python
# Integrate the following differential equations
# dx/dt = x - x^3 - y
# dy/dt = -a*y + x
# with initial conditions x(0) = 0.1, y(0) = 0.1

# First plot the nullclines

import numpy as np
import matplotlib.pyplot as plt

def dxdt(x, y, I):
    return x - x**3 - y - I

def dydt(x, y, a):
    return -a*y + x

def x_nullcline(x, a):
    return x-x**3 - I

def y_nullcline(y, a):
    return 1/a*x

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

I = -0.5
a = 1

plt.plot(x, x_nullcline(x, a), label='x nullcline')
plt.plot(y, y_nullcline(y, a), label='y nullcline')
plt.legend()
plt.show()
```

```python
#quiver plot over the nullclines -- downsampled for clarity

xd = np.linspace(-1, 1, 10)
yd = np.linspace(-1, 1, 10)

X, Y = np.meshgrid(xd, yd)

U = dxdt(X, Y, a)
V = dydt(X, Y, a)

plt.quiver(X, Y, U, V)
plt.plot(x, x_nullcline(x, a), label='x nullcline')
plt.plot(y, y_nullcline(y, a), label='y nullcline')
plt.legend()
plt.show()
```

```python
from scipy.integrate import odeint

def dXdt(X, t, I, a):
    x, y = X
    return [dxdt(x, y, I), dydt(x, y, a)]

t = np.linspace(0, 50, 20000)

X0 = [1, 1]

sol = odeint(dXdt, X0, t, args=(I,a))
sol_x, sol_y = sol.T

plt.quiver(X, Y, U, V)
plt.plot(x, x_nullcline(x, a), label='x nullcline')
plt.plot(y, y_nullcline(y, a), label='y nullcline')
plt.legend()
plt.plot(sol_x,sol_y)
plt.show()
```
