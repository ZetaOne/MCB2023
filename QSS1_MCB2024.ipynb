```python
#import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dAdt (k1,A):
    return -k1*A

def dBdt (k1,k2,A,B):
    return k1*A - k2*B

def dCdt (k2,B):
    return k2*B
```

```python
#initial conditions
A0 = 0.7
B0 = 0.2
C0 = 0.1

#rate constants
k1 = 10
k2 = 1
```

```python
######## integrating using Euler's method ########

dt = 1e-4

tstart = 0
tend = 15

t = tstart

#make empty lists to store the values of A, B, and C
A_list = []
B_list = []
C_list = []
t_list = []
A_list.append(A0)
B_list.append(B0)
C_list.append(C0)
t_list.append(t)

while t < tend:


    A0 = A0 + dAdt(k1,A0)*dt
    B0 = B0 + dBdt(k1,k2,A0,B0)*dt
    C0 = C0 + dCdt(k2,B0)*dt

    t = t + dt

    A_list.append(A0)
    B_list.append(B0)
    C_list.append(C0)
    t_list.append(t)
```

```python
#plot A , B and C on the same graph as subplots
figure = plt.figure(figsize=(10,8),dpi=120)
plt.subplot(3,1,1)
plt.plot(t_list,A_list,label='A', color='red')
plt.xlim(-0.10,)
plt.legend()
plt.legend()
plt.subplot(3,1,2)
plt.plot(t_list,B_list,label='B', color='green')
plt.xlim(-0.10,)
plt.legend()
plt.subplot(3,1,3)
plt.plot(t_list,C_list,label='C', color='blue')
plt.xlim(-0.10,)
plt.xlabel('Time')
plt.legend()
plt.show()
```
